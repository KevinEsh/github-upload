```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```
data = pd.read_csv("gov.csv", header=None, encoding='utf-8')
data
```

```
name = data.iloc[0,4]
colors = ['red', 'blue', 'green', 'yellow','pink', 'lightgreen', 'lightblue', 'wheat', 'salmon', 'brown']
beers = [data.iloc[0,0]]
c = 0
start, end = 0, 1
fig=plt.figure(figsize=(25,10))
ax = fig.add_subplot(111)

for i in range(data.shape[0]):
    #name = data.iloc[i,0]
    if name != data.iloc[i,4]:
        name = data.iloc[i,4]
        start +=1
        end += 1
        c = (c+1) % len(colors)
        beers.append(data.iloc[i,0])

    plt.fill_between([data.iloc[i,2], data.iloc[i,3] - 10], [start, start], [end,end], color=colors[c], edgecolor='k', linewidth=0.5)
    plt.text((data.iloc[i,2]+data.iloc[i,3])/2 - 30, start+0.3 , data.iloc[i,1], fontsize=12)

minor_ticks = np.arange(0, np.max(data.iloc[:,3]) +1,120)
ax.set_xticks(minor_ticks)
ax.xaxis.grid()
ax.set_yticks(np.arange(0.5, start+1, 1.0))
ax.set_yticklabels(beers)
plt.show()
```
