```
import pandas as pd
```

```
def date_to_int(date):
    return str(60*24*(date[0]-1) + 60*date[1] + date[2])
```

```
dataframe = pd.read_csv("plan_gob_2020-02-01.csv")
dataframe.head()
```

```
with open('query.txt', 'w') as file:
    suma = 0
    ordenes = dataframe.ordenes.unique()
    tanks = dataframe.tanque.unique()
    file.write(str(ordenes.shape[0]) + " " + str(tanks.shape[0]) + "\n")

    for orden in ordenes:
        temp = dataframe.loc[dataframe['ordenes'] == orden].iloc[0].loc[["marca", "inicio liquido", "fin liquido", "HL liquido"]]
        marca = str(temp[0].replace(' ', '_'))
        start = date_to_int(list(map(int, [temp[1].split('/')[2].split(' ')[0]] + temp[1].split()[1].split(':') ) ) )
        end = date_to_int(list(map(int, [temp[2].split('/')[2].split(' ')[0]] + temp[2].split()[1].split(':') ) ) )
        amount = str(temp[3])
        suma += temp[3]
        print(marca, start, end, amount)
        file.write(marca+" "+start+" "+end+" "+amount+"\n")

    for tank in tanks:
        print(tank, 1550)
        file.write(tank + " "+ str(1550)+ "\n")

    print(suma, suma/tanks.shape[0])
```
